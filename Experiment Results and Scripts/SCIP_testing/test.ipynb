```python
import re
import numpy as np
import pandas as pd
for dataset in ['auslan2','aslbu','pioneer','context','robot','epitope','skating','question','reuters','gene','unix']:
    # 定义文件路径
    file_path = f'Python_{dataset}.txt_SCIS.out'

    # 初始化列表来存储准确率
    acc_har = []
    acc_match = []

    # 正则表达式模式来找到准确率
    pattern_har = re.compile(r'Accuracy_HAR \d+:(\d+\.\d+)')
    pattern_match = re.compile(r'Accuracy_MATCH \( topk\d+\)\d+:(\d+\.\d+)')

    # 读取文件并提取准确率
    with open(file_path, 'r') as file:
        for line in file:
            match_har = pattern_har.search(line)
            match_match = pattern_match.search(line)
            if match_har:
                acc_har.append(float(match_har.group(1)))
            if match_match:
                acc_match.append(float(match_match.group(1)))

    # 计算每个算法的均值和标准差
    mean_har = np.mean(acc_har)/100
    std_har = np.std(acc_har)/100
    mean_match = np.mean(acc_match)/100
    std_match = np.std(acc_match)/100

    # 创建结果字典
    mean_std_results = {
        'Algorithm': ['HAR', 'MATCH'],
        'Mean Accuracy': [mean_har, mean_match],
        'Standard Deviation': [std_har, std_match]
    }

    # 转换为DataFrame
    results_df = pd.DataFrame(mean_std_results)
    # save to csv
    results_df.to_csv(f'./{dataset}_mean_std_SCIS.csv', index=False)
```
